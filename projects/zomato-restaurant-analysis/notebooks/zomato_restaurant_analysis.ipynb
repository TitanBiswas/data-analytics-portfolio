- res_id – A unique ID number for the restaurant

- establishment – The type of outlet (e.g., Quick Bites, Casual
Dining)
- url – Link to the restaurant’s Zomato webpag
- address – Full physical address of the restaurant
- city – City where the restaurant is located
- city_id – Zomato’s internal ID for the city
- locality – The specific area or neighborhood
- latitude – GPS latitude coordinate
- longitude – GPS longitude coordinate
- zipcode – The postal code (may be missing)
- country_id – Zomato’s internal ID for the country
- locality_verbose – Detailed locality description
- cuisines – The types of cuisines served (e.g., North Indian,

```python
def check(input_list):
    if 'Delivery' in input_list:
        return 'Delivery Available'
    elif 'Takeaway Available' in input_list:
        return 'Takeaway Available'
    else:
        return 'No Delivery Available'
```

```python
ff=df['highlights'].apply(check).value_counts().reset_index()
ff
```

```python
plt.pie(ff['count'],labels=ff['highlights'],autopct='%.1f%%')
plt.title('Restaurants Service Availablity',fontsize=18,fontweight='bold')
plt.show()
```
