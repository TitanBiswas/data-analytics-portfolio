# Manupelating the Data (Filling Null Values & Outliers/Incorrect)

```python
# data.drop['index', axis = 1,inplace=True]
```

```python
# Filling the Missing Values in Product, Brand, Description

data['product'] = data['product'].fillna('unknown')
data['brand'] = data['brand'].fillna('unknown')
data['description'] = data['description'].fillna('unknown')
# data['sale_price'] = data['sale_price'].fillna('unknown')
```

```python
# Filling the Missing Values in Sale price with market Price

data.sale_price = data.sale_price.fillna(data.market_price)
```

```python
# Filling the Missing Ratings with Mean

data.rating= data['rating'].fillna(data.rating.mean())
```

```python
data.isnull().sum()
```

```python
data.drop('index',axis=1,inplace=True)
```

```python
data.head()
```

```python
data[data['sale_price'] > data['market_price']]
```

```python
# Here filled the Market price where sale price is greater than market Price

data.sale_price = np.where(data.sale_price > data.market_price,data.market_price,data.sale_price)
```

```python
data['discount'] = data['market_price'] - data['sale_price']
```

```python
data['discount_per'] = (data['discount'] / data['market_price']) * 100
```

# Detecting & Treating Outliers

```python
# Sale Price

sns.boxplot(data.sale_price,orient='h')
```

```python
# IQR METHOD
# Step 1 : Finding Q1 and Q3....

Q1,Q3 = data.sale_price.quantile([0.25,0.75])
```

```python
Q1,Q3
```

```python
# Step 2 : Finding IQR 

IQR = Q3 - Q1
```

```python
IQR
```

```python
# Step 3 Finding Upper Limit and Lower Limit...

UL = Q3 + 1.5*(IQR)
LL = Q1 - 1.5*(IQR)
```

```python
UL,LL
```

```python
# data[data.sale_price>UL].sort_values(data.sale_price,ascending=False).head()
data[data.sale_price > UL].sort_values('sale_price', ascending=False)
```
